```r
# === Phylogenetic Independent Contrasts: All Volvocine Species ===

library(ape)

obj <- read.csv("volvocine_multicellular_with_gonidia.csv", row.names = 1)
vtree <- read.tree("volvocine tree3 newick.txt")

max.cells <- setNames(obj[, "max_cells"], rownames(obj))
pgerm.max <- setNames(obj[, "p_germ_max"], rownames(obj))
logmax.cells <- log(max.cells)
logpgerm.max <- log(pgerm.max)

pic.lmc <- pic(logmax.cells, vtree)
pic.lpgm <- pic(logpgerm.max, vtree)

fit.pic <- lm(pic.lpgm ~ pic.lmc)
summary(fit.pic)


# === Phylogenetic Independent Contrasts: Volvox Only ===

obj <- read.csv("volvox_multicellular_with_gonidia.csv", row.names = 1)
vtree <- read.tree("volvox only tree newick.txt")

max.cells <- setNames(obj[, "max_cells"], rownames(obj))
pgerm.max <- setNames(obj[, "p_germ_max"], rownames(obj))
logmax.cells <- log(max.cells)
logpgerm.max <- log(pgerm.max)

pic.lmc <- pic(logmax.cells, vtree)
pic.lpgm <- pic(logpgerm.max, vtree)

fit.pic <- lm(pic.lpgm ~ pic.lmc)
summary(fit.pic)
```
